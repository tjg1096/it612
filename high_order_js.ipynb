```python
%%writefile code.js

// PART A
const data = [5, 12, 3, 20, 11, 9, 30];

// Loop
const loopResult = [];
for (let i = 0; i < data.length; i++) {
  if (data[i] > 10) {
    loopResult.push(data[i] * 2);
  }
}

console.log("loopResult:", loopResult);

// Pipeline
const pipelineResult = data
  .filter(n => n > 10)
  .map(n => n * 2);

console.log("pipelineResult:", pipelineResult);

console.log(
  "Match?",
  JSON.stringify(loopResult) === JSON.stringify(pipelineResult)
);

// PART B
const numbers = [1, 2, 3, 4, 5, 6];

const sumDoubledEvens = numbers
  .filter(n => n % 2 === 0)
  .map(n => n * 2)
  .reduce((sum, n) => sum + n, 0);

console.log("sumDoubledEvens:", sumDoubledEvens);

// PART C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

const anyUnder18 = users.some(user => user.age < 18);
console.log("anyUnder18:", anyUnder18);

const userId3 = users.find(user => user.id === 3);
console.log("userId3:", userId3);

console.log("Names:");
users.forEach(user => console.log(user.name));

// PART D
function repeat(n, action) {
  for (let i = 0; i < n; i++) {
    action(i);
  }
}

console.log("Test 1:");
repeat(3, console.log);

console.log("Test 2:");
repeat(4, i => console.log("Step " + i));
```
